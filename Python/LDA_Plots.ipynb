```python
import numpy as np

# Create the vectors X and Y
x = np.arange(0.5, 3.25, 0.01, dtype=np.double)
y = -0.4*x+1

# Create the plot
plt.plot(x,y, 'k', label='line 0')


y1=[0.6, 0.5, 0.48, 0.47, 0.46, 0.45]
y2=[0.2, 0.25, 0.22, 0.1, 0.18, 0]

my1=np.mean(y1)
my2=np.mean(y2)
print(my1)
print(my2)

x1=[]
for y in y1:
    x1.append((-y+1)/(0.4))
x2=[]
for y in y2:
    x2.append((-y+1)/(0.4))
plt.plot(x1, y1, 'ro', label='line 2')
plt.plot(x2, y2, 'go', label='line 1')
plt.plot((-my1+1)/(0.4), my1, 'r+', markersize=30)
plt.plot((-my2+1)/(0.4), my2, 'g+', markersize=30)

```
