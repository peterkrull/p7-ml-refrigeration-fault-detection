```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
```

```python
x = np.arange(0, 1000, 0.1, dtype=np.double)
y1=-x
y2=-x+200
y3=-x+400
plt.plot(x, y1, 'k', label='line 0')
plt.plot(x, y2, 'k--', label='line 1')
plt.plot(x, y3, 'k', label='line 2')

px=200
plt.plot(px, -px, 'go', markersize=5)
plt.plot(px, -px+400, 'ro', markersize=5)
px=700
plt.plot(px, -px+400, 'ro', markersize=5)


i=0
x_p1=[]
y_p1=[]
while(i<100):
    y_p_1 = random.uniform(np.min(y3), np.max(y3))
    x_p_1 = random.uniform(np.min(x), np.max(x))
    if(y_p_1<-x_p_1-20):
        x_p1.append(x_p_1)
        y_p1.append(y_p_1)
        i=i+1


i=0
x_p2=[]
y_p2=[]
while(i<100):
    y_p_2 = random.uniform(np.min(y1), np.max(y1))
    x_p_2 = random.uniform(np.min(x), np.max(x))
    if(y_p_2>-x_p_2+420):
        x_p2.append(x_p_2)
        y_p2.append(y_p_2)
        i=i+1

plt.plot(x_p1, y_p1, 'go', label='line 3')
plt.plot(x_p2, y_p2, 'ro', label='line 4')
```

```python
from shapely.geometry import Point, Polygon
def Random_Points_in_Polygon(polygon, number):
    points = []
    minx, miny, maxx, maxy = polygon.bounds
    while len(points) < number:
        pnt = Point(np.random.uniform(minx, maxx), np.random.uniform(miny, maxy))
        if polygon.contains(pnt):
            points.append(pnt)
    return points

x = np.arange(0, 1000, 0.1, dtype=np.double)
y1=-x
y2=-x+200
y3=-x+400
plt.plot(x, y1, 'k', label='line 0')
plt.plot(x, y2, 'k--', label='line 1')
plt.plot(x, y3, 'k', label='line 2')

px=200
plt.plot(px, -px, 'go', markersize=5)
plt.plot(px, -px+400, 'ro', markersize=5)
px=700
plt.plot(px, -px+400, 'ro', markersize=5)
```
