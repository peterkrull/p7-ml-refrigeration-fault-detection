```python
from shapely.geometry import Point, Polygon
def Random_Points_in_Polygon(polygon, number):
    points = []
    minx, miny, maxx, maxy = polygon.bounds
    while len(points) < number:
        pnt = Point(np.random.uniform(minx, maxx), np.random.uniform(miny, maxy))
        if polygon.contains(pnt):
            points.append(pnt)
    return points

x = np.arange(0, 1000, 0.1, dtype=np.double)
y1=-x
y3=-x+400
plt.plot(x, y1, 'k')
plt.plot(x, y3, 'k')

u=40
polygon = Polygon([[min(x+u), max(y3+u)],[max(x-u), max(y3+u)],[max(x-u), min(y3+u)],[min(x+u), max(y3+u)]])
points = Random_Points_in_Polygon(polygon, 100)

xs = [point.x for point in points]
ys = [point.y for point in points]

plt.plot(xs, ys, 'rs')



polygon2 = Polygon([[min(x+u), max(y1-u)],[min(x+u), min(y1-u)],[max(x-u), min(y1-u)],[min(x+u), max(y1-u)]])
points2 = Random_Points_in_Polygon(polygon2, 100)
xs2 = [point2.x for point2 in points2]
ys2 = [point2.y for point2 in points2]

plt.plot(xs2, ys2, 'gs')


pu=250
#plt.arrow(pu, -pu, 100, 300, head_width = 20, width = 0.05, ec ='green')
plt.arrow(pu, -pu, 80, 280, head_width = 20, width = 0.05, ec ='green')
pu=600
plt.arrow(pu, -pu, 80, 280, head_width = 20, width = 0.05, ec ='green')
#xu2=pu
#yu2=-pu+400
#plt.arrow(pu, -pu, -pu+xu2, pu+yu2, head_width = 40, width = 0.05, ec ='green')
```
