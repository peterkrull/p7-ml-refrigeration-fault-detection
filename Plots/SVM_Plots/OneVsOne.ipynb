```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
     "text.usetex": True,
     "font.family": "cmr"
 }) 
```

```python
#Plots the lines 
x = np.arange(-10, 30, 0.01, dtype=np.float64)
y1=48-0.1*x
y2=93-7*x
y3=-0.731+2*x

plt.plot(x, y1, 'C0',label='A VS B')
plt.plot(x, y2, 'C1',label='B VS C')
plt.plot(x, y3, 'C2',label='A VS C')

#Plot the point that is which to determine the class for 
px=2.5
py=60

plt.plot(px, py, 'C3.', markersize=30,label='Text Point')


#Text plots: 
front=35
plt.text(20, 0, 'C', fontsize = front)

plt.text(0, 20, 'B', fontsize = front)

plt.text(10, 65, 'A', fontsize = front)

plt.text(12, 32.5, '?', fontsize = front)

#Something abouting plotting
plt.axis('off')
#plt.grid('on')
plt.xlim([-5,30])
plt.ylim([-20,80])
plt.legend(loc='upper right')
plt.savefig("OneVsOne.pdf", format="pdf", bbox_inches="tight")
#plt.axis('equal')
```
