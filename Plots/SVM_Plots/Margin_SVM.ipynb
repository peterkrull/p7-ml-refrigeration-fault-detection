```python
#pip install shapely 
import matplotlib.pyplot as plt
import numpy as np
from shapely.geometry import Point, Polygon

plt.rcParams.update({
     "text.usetex": True,
     "font.family": "cmr"
 })
```

```python
#Something to plot/makes the points for the two classes!
def Random_Points_in_Polygon(polygon, number):
    points = []
    minx, miny, maxx, maxy = polygon.bounds
    while len(points) < number:
        pnt = Point(np.random.uniform(minx, maxx), np.random.uniform(miny, maxy))
        if polygon.contains(pnt):
            points.append(pnt)
    return points

#Plots Decision boundary and the two other lines
x = np.arange(0, 1000, 0.1, dtype=np.double)
y1=-x
y2=-x+200
y3=-x+400

plt.plot(x, y1, 'gray')
plt.plot(x, y2, 'k',label='Decision Boundary')
plt.plot(x, y3, 'gray',label='Support Vector Line')


#Support vector points 
""""
px=200
plt.plot(px, -px, 'C1.', markersize=10)
px=300
plt.plot(px, -px+400, 'C0.', markersize=10)
px=700
plt.plot(px, -px+400, 'C0.', markersize=10)
"""
px=200

plt.plot(px, -px, 'C1.')

plt.scatter(px, -px, s=70, facecolors='none', edgecolors='C2',label='Support Vector')

px=300
plt.plot(px, -px+400, 'C0.')

plt.scatter(px, -px+400, s=70, facecolors='none', edgecolors='C2')

px=700

plt.plot(px, -px+400, 'C0.')

plt.scatter(px, -px+400, s=70, facecolors='none', edgecolors='C2')

#Data points plot
n=150 #number of data points that are plotted
u=20
polygon = Polygon([[min(x), max(y1)-u],[max(x)-u, min(y1)],[min(x), min(y1)],[min(x), max(y1)-u]])

points = Random_Points_in_Polygon(polygon, n) 

xs = [point.x for point in points]
ys = [point.y for point in points]

plt.plot(xs, ys, 'C1.', label='Class 1')

polygon2 = Polygon([[min(x+u), max(y3)], [max(x), max(y3)],[max(x), min(y3+u)],[min(x+u), max(y3)]])

points2 = Random_Points_in_Polygon(polygon2, n)

xs2 = [point2.x for point2 in points2]
ys2 = [point2.y for point2 in points2]

#Boxes used for Margin Support vector and decision boundary 

plt.plot(xs2, ys2, 'C0.', label='Class 2')
plt.text(110, -10, 'Margin', fontsize = 10, bbox = dict(facecolor = 'white', alpha = 0.9))



#plt.text(550, 100, 'Support Vectors', fontsize = 10, bbox = dict(facecolor = 'white', alpha = 0.9), zorder = 9)


#plt.text(450, -350, 'Decision\nBoundary', fontsize = 10, bbox = dict(facecolor = 'white', alpha = 0.9))

#Arrow for margin:

import matplotlib.patches as patches
style = "Simple, tail_width=0.5, head_width=4, head_length=8"
kw = dict(arrowstyle=style, color="k")
# a1 = patches.FancyArrowPatch((100, 0), (150, 50), **kw)
a1 = patches.FancyArrowPatch((95, 110), (50, -50), arrowstyle='<->', mutation_scale=10)
plt.gca().add_patch(a1)
#a2 = patches.FancyArrowPatch((100, 0), (50, -50), **kw)
#plt.gca().add_patch(a2)
#plt.annotate(s='', xy=(1,1), xytext=(0,0), arrowprops=dict(arrowstyle='<->'))

#Arrows for support vectors
""""
px=200
a3 = patches.FancyArrowPatch((700, 120), (px, -px), **kw, zorder = 8)
plt.gca().add_patch(a3)
px=300
a4 = patches.FancyArrowPatch((700, 120), (px, -px+400), **kw, zorder = 8)
plt.gca().add_patch(a4)
px=700
a5 = patches.FancyArrowPatch((700, 120), (px, -px+400), **kw, zorder = 8)
plt.gca().add_patch(a5)
"""

#Something abouting plotting
plt.axis('off')
plt.legend(loc='upper right')
plt.savefig("Margin_SVM.pdf", format="pdf", bbox_inches="tight")
#plt.axis('equal')
#200, -200
#200, 200
#700, -300
```
